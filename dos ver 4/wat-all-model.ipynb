```python
import pandas as pd
from sklearn import preprocessing
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.decomposition import PCA
```

```python
train_df = pd.read_csv("wat-all.csv")
```

```python
train_df.head()
```

```python
train_X = train_df.drop(columns=['packet_address','time','target',"packet_address_id"])
```

```python
train_X
```

```python
x = train_X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
train_X = pd.DataFrame(x_scaled)
train_X
```

```python
corr_df = pd.concat([train_X, train_df[['target']]], axis = 1)
corr_df.corr()
```

```python
train_Y = train_df['target']
```

```python
train_Y
```

```python
model = Sequential()
```

```python
n_cols = train_X.shape[1]
n_cols
```

```python
model.add(Dense(32, activation='relu', input_shape=(n_cols,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
early_stopping_monitor = EarlyStopping(patience=20)
```

```python
pca = PCA(0.95)
```

```python
pca.fit(train_X)
```

```python
principal_components = pca.transform(train_X)
principal_components
```

```python
pca.explained_variance_ratio_
```

```python
features = range(pca.n_components_)
```

```python
plt.bar(features, pca.explained_variance_)
plt.xticks(features)
plt.xlabel("PCs")
plt.ylabel("Variance")
```

```python
principal_df = pd.DataFrame(data = principal_components)
principal_df
```

```python
final_df = pd.concat([principal_df, train_df[['target']]], axis = 1)
final_df
```

```python
final_df.corr()
```

```python
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2D PCA', fontsize = 20)
targets = [1, 0]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = final_df['target'] == target
    ax.scatter(final_df.loc[indicesToKeep, 0]
               , final_df.loc[indicesToKeep, 1]
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
model.fit(train_X, train_Y, epochs=30, validation_split=0.2, callbacks=[early_stopping_monitor])
```
