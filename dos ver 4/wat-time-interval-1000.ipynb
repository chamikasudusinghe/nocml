```python
#convert txt to csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfg = pd.read_csv('wat-good-time-interval-1000.csv')
```

```python
dfm = pd.read_csv('wat-mal-time-interval-1000.csv')
```

```python
dfg.head()
```

```python
dfm.head()
```

```python
df = dfg.append(dfm, ignore_index=True)
```

```python
df.head()
```

```python
df = df.sort_values('time_intervals')
```

```python
df.head(50)
```

```python
df['tot_packets']= df.iloc[:, 1:16].sum(axis=1)
```

```python
df['tot_mean']= df.iloc[:, 1:16].mean(axis=1)
```

```python
df
```

```python
df.to_csv('wat-time-interval-1000.csv',index=False)
```
