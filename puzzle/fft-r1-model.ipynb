```python
import time
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

Importing Dataset from GitHub

```python
df = pd.read_csv('results_tiles_manhattan.csv')
dfms = pd.read_csv('results_mean_std.csv')
```

```python
x = dfms['n']
y1 = dfms['mean_tiles']
y2 = dfms['mean_manhatan']


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

plt.title("mean time")
ax1.set_xlabel('n-puzzle')
ax1.set_ylabel('tile time (ms)', color='g')
ax2.set_ylabel('manhatan time (ms)', color='b')

plt.show()
```

```python
fig,ax=plt.subplots()
plt.title("mean time")
ax.set_xlabel("n-puzzle")
ax.set_ylabel("time (ms)")
ax.plot(x, y1, marker="o")
ax.plot(x, y2, marker="o")
plt.show()
```

```python
y3 = dfms['std_tiles']
y4 = dfms['std_manhatan']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y3, 'g-')
ax2.plot(x, y4, 'b-')

plt.title("standard deviation")
ax1.set_xlabel('n-puzzle')
ax1.set_ylabel('tile time (ms)', color='g')
ax2.set_ylabel('manhatan time (ms)', color='b')

plt.show()
```

```python
fig,ax=plt.subplots()
plt.title("standard deviation")
ax.set_xlabel("n-puzzle")
ax.set_ylabel("time (ms)")
ax.plot(x, y3, marker="o")
ax.plot(x, y4, marker="o")
plt.show()
```
