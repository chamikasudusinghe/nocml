```python
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wat-good.csv')
df = df.loc[df['router'] == 1]
df = df.drop(columns=['router'])
df.to_csv('wat-r1-good.csv',index=False)
```

```python
df = pd.read_csv('wat-r1-good.csv')
df
```

```python
timearr = []
interval = 99
count = 0
for index, row in df.iterrows():
    if row["time"]<=interval:
        count+=1
    else:
        timearr.append([interval+1,count])
        count=1
        interval+=100
timearr.append([interval+1,count])
```

```python
countarr = []
maxarr = []
for i in range(len(timearr)):
    for cnt in range(timearr[i][1],0,-1):
        countarr.append(cnt)
        maxarr.append(timearr[i][1])
print(len(countarr))
```

```python
df = df.assign(pkt_count=countarr)
df = df.assign(pkt_max=maxarr)
```

```python
cols = list(df.columns)
a, b = cols.index('target'), cols.index('pkt_count')
cols[b], cols[a] = cols[a], cols[b]
df = df[cols]
df
cols = list(df.columns)
a, b = cols.index('target'), cols.index('pkt_max')
cols[b], cols[a] = cols[a], cols[b]
df = df[cols]
df
```
